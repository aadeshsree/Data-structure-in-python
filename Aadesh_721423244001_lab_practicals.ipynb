1.Write a program for list operations.

```python
#Creating a list
n=int(input("Enter the number of items:"))
l=[]
for i in range(0,n):
    a=int(input("Enter the element:"))
    l.append(a)
#inserting element in a list
a=int(input("Enter the position to insert:"))
b=int(input("Enter the value:"))
l.insert(a,b)
print("The list with inserted element",l)
#deleting element in a list
d=int(input("Enter the position to delete:"))
del(l[d])
print("The list with deleted element",l)
#Travesing a list
for i in range(len(l)):
    print(l[i])
```

2.Write a program to accept 4 digit number and sum all the numbers.

3.Write an program to accept 10 integers and sum 10 integers in an array.

```python
n = input("Enter the values using commas: ")
v = n.split(",")
A=[]
sum=0
for i in v:
    A.append(int(i))
if len(A)!=10:
    print("Please enter 10 digits")
else:
    for i in A:
        sum=sum+i
    print(sum)
```
